```python
import pandas as pd
from datetime import datetime, timedelta
import pytz

# Initialize data structures
mentors = {}
students = {}
bookings = pd.DataFrame(columns=['mentor_id', 'student_id', 'slot_time', 'duration', 'status', 'doubt_description'])
timezone = pytz.timezone('UTC')

class Mentor:
    def __init__(self, mentor_id, name, expertise, available_slots=None):
        self.id = mentor_id
        self.name = name
        self.expertise = expertise
        self.available_slots = available_slots if available_slots else []
        mentors[mentor_id] = self

    def add_availability(self, start_time, end_time, duration_minutes=30):
        current_time = start_time
        while current_time + timedelta(minutes=duration_minutes) <= end_time:
            self.available_slots.append({
                'start': current_time,
                'end': current_time + timedelta(minutes=duration_minutes),
                'booked': False
            })
            current_time += timedelta(minutes=duration_minutes)

    def view_bookings(self):
        return bookings[bookings['mentor_id'] == self.id].copy()

    def mark_doubt_cleared(self, booking_id):
        if booking_id in bookings.index:
            bookings.loc[booking_id, 'status'] = 'Completed'
            return True
        return False

class Student:
    def __init__(self, student_id, name):
        self.id = student_id
        self.name = name
        students[student_id] = self

    def view_available_slots(self, mentor_id=None):
        if mentor_id:
            if mentor_id in mentors:
                mentor = mentors[mentor_id]
                return [slot for slot in mentor.available_slots if not slot['booked']]
            return []
        else:
            all_slots = []
            for mentor in mentors.values():
                all_slots.extend([(mentor.id, mentor.name, slot)
                                for slot in mentor.available_slots if not slot['booked']])
            return all_slots

    def book_slot(self, mentor_id, slot_index, doubt_description):
        if mentor_id not in mentors:
            return False, "❌ Mentor not found"

        mentor = mentors[mentor_id]
        available_slots = [i for i, slot in enumerate(mentor.available_slots) if not slot['booked']]

        if slot_index >= len(available_slots):
            return False, "❌ Invalid slot index"

        actual_index = available_slots[slot_index]
        slot = mentor.available_slots[actual_index]

        if slot['booked']:
            return False, "❌ Slot already booked"

        slot['booked'] = True
        new_booking = {
            'mentor_id': mentor_id,
            'student_id': self.id,
            'slot_time': slot['start'],
            'duration': (slot['end'] - slot['start']).seconds // 60,
            'status': 'Scheduled',
            'doubt_description': doubt_description
        }

        global bookings
        bookings = pd.concat([bookings, pd.DataFrame([new_booking])], ignore_index=True, sort=False)
        return True, "✅ Slot booked successfully!"

    def view_my_bookings(self):
        return bookings[bookings['student_id'] == self.id].copy()

def setup_demo_data():
    global bookings
    mentors.clear()
    students.clear()
    bookings = pd.DataFrame(columns=['mentor_id', 'student_id', 'slot_time', 'duration', 'status', 'doubt_description'])

    # Create mentors with sample availability
    mentor1 = Mentor('m1', 'Dr. Smith', ['Python', 'Machine Learning'])
    mentor2 = Mentor('m2', 'Prof. Johnson', ['Data Structures', 'Algorithms'])

    now = datetime.now(timezone).replace(minute=0, second=0, microsecond=0)

    # Set up mentor availability (next 3 days)
    mentor1.add_availability(now + timedelta(days=1, hours=10), now + timedelta(days=1, hours=12))
    mentor1.add_availability(now + timedelta(days=2, hours=14), now + timedelta(days=2, hours=16))
    mentor2.add_availability(now + timedelta(days=1, hours=9), now + timedelta(days=1, hours=11))
    mentor2.add_availability(now + timedelta(days=3, hours=13), now + timedelta(days=3, hours=15))

    # Create students
    Student('s1', 'Alice')
    Student('s2', 'Bob')
    Student('s3', 'Charlie')

def display_available_slots():
    print("\n📅 Available Slots:")
    for mentor_id, mentor in mentors.items():
        print(f"\n👨‍🏫 Mentor: {mentor.name} ({', '.join(mentor.expertise)})")
        available_slots = [slot for slot in mentor.available_slots if not slot['booked']]
        if not available_slots:
            print("  ⚠ No available slots")
            continue
        for i, slot in enumerate(available_slots):
            print(f"  {i}: 🕒 {slot['start'].strftime('%a, %b %d %H:%M')} to {slot['end'].strftime('%H:%M')}")

def display_bookings(df):
    if df.empty:
        print("📭 No bookings found")
    else:
        print(df.to_string(index=True))

def main():
    setup_demo_data()
    print("\n" + "="*50)
    print("🌟 Welcome to Mentor Slot Booking System 🌟")
    print("="*50)

    while True:
        print("\n🔹 Main Menu:")
        print("1. View available slots")
        print("2. Book a slot (student)")
        print("3. View my bookings (student)")
        print("4. View mentor bookings")
        print("5. Mark doubt as cleared (mentor)")
        print("6. Reset demo data")
        print("7. Exit")

        choice = input("\n👉 Enter your choice (1-7): ").strip()

        if choice == '1':
            display_available_slots()

        elif choice == '2':
            print("\n🎯 Book a Slot:")
            student_id = input("Enter your student ID (s1/s2/s3): ").strip().lower()
            if student_id not in students:
                print("❌ Invalid student ID")
                continue

            display_available_slots()
            mentor_id = input("\nEnter mentor ID (m1/m2): ").strip().lower()
            if mentor_id not in mentors:
                print("❌ Invalid mentor ID")
                continue

            try:
                slot_index = int(input("Enter slot number: "))
                doubt = input("Describe your doubt/question: ")
                success, msg = students[student_id].book_slot(mentor_id, slot_index, doubt)
                print(msg)
            except ValueError:
                print("❌ Please enter a valid number")

        elif choice == '3':
            print("\n📖 My Bookings:")
            student_id = input("Enter student ID (s1/s2/s3): ").strip().lower()
            if student_id not in students:
                print("❌ Invalid student ID")
                continue
            display_bookings(students[student_id].view_my_bookings())

        elif choice == '4':
            print("\n📊 Mentor Bookings:")
            mentor_id = input("Enter mentor ID (m1/m2): ").strip().lower()
            if mentor_id not in mentors:
                print("❌ Invalid mentor ID")
                continue
            display_bookings(mentors[mentor_id].view_bookings())

        elif choice == '5':
            print("\n✅ Clear Doubt:")
            mentor_id = input("Enter mentor ID (m1/m2): ").strip().lower()
            if mentor_id not in mentors:
                print("❌ Invalid mentor ID")
                continue

            mentor_bookings = mentors[mentor_id].view_bookings()
            display_bookings(mentor_bookings)

            if not mentor_bookings.empty:
                try:
                    booking_id = int(input("Enter booking index to mark as completed: "))
                    if booking_id not in mentor_bookings.index:
                        print("❌ Invalid booking ID")
                        continue

                    if mentors[mentor_id].mark_doubt_cleared(booking_id):
                        print("✅ Doubt marked as cleared!")
                    else:
                        print("❌ Failed to update booking")
                except ValueError:
                    print("❌ Please enter a valid number")

        elif choice == '6':
            setup_demo_data()
            print("\n🔄 Demo data has been reset!")

        elif choice == '7':
            print("\n👋 Thank you for using the Mentor Slot Booking System!")
            break

        else:
            print("❌ Invalid choice. Please enter a number between 1-7.")

# Run the program
if __name__ == "__main__":
    main()
```

```python
import os
import random
from langchain_community.document_loaders import PyPDFLoader, TextLoader, Docx2txtLoader
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from gtts import gTTS

# 🔧 Configurations
CHROMA_DIR = "SB"  # Directory for storing the Chroma vector DB
EMBED_MODEL = "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"  # Embedding model for multilingual text
DOCUMENTS_DIR = "documents"  # Directory containing your documents

# 📚 Load document based on file type
def load_file(filepath):
    ext = filepath.split('.')[-1].lower()
    if ext == "pdf":
        return PyPDFLoader(filepath).load()
    elif ext == "txt":
        return TextLoader(filepath, encoding="utf-8").load()
    elif ext == "docx":
        return Docx2txtLoader(filepath).load()
    else:
        raise ValueError("Unsupported file format!")

# 🧩 Split document into smaller chunks (for efficient embedding and storage)
def chunk_documents(docs, chunk_size=500, chunk_overlap=50):
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap
    )
    return splitter.split_documents(docs)

# 🔮 Embed document chunks and store them in the Chroma vector store
def embed_and_save(chunks, persist_dir=CHROMA_DIR):
    embedding = HuggingFaceEmbeddings(model_name=EMBED_MODEL)  # Load the embedding model
    vectordb = Chroma.from_documents(documents=chunks, embedding=embedding, persist_directory=persist_dir)
    vectordb.persist()  # Persist the vector store to disk
    print(f"✅ Successfully saved {len(chunks)} chunks to vector store at: {persist_dir}")

# 📖 Process the file: Load, chunk, embed, and save
def process_file(file_path):
    print(f"📖 Loading file: {file_path}")
    docs = load_file(file_path)  # Load document
    print(f"📄 Loaded {len(docs)} document(s). Splitting into chunks...")
    chunks = chunk_documents(docs)  # Split the document into chunks
    print(f"🧩 Created {len(chunks)} chunks. Saving to DB...")
    embed_and_save(chunks)  # Embed and save to Chroma DB
    print("🎉 Done!")

# 🔊 Text-to-Speech (TTS) function to convert text to speech (audio)
def text_to_audio(text, lang='en', filename='output.mp3'):
    try:
        tts = gTTS(text=text, lang=lang)  # Convert text to speech
        tts.save(filename)  # Save audio file
        print(f"✅ Audio saved to {filename}")
        # Auto-play the audio (for Windows OS)
        os.system(f"start {filename}")
    except Exception as e:
        print(f"❌ Error: {e}")

# 📂 Main function to execute document processing and TTS
if __name__ == "__main__":
    # Randomly select a file from the 'documents' directory
    try:
        files = [f for f in os.listdir(DOCUMENTS_DIR) if f.endswith(('.pdf', '.txt', '.docx'))]
        if files:
            random_file = random.choice(files)  # Randomly choose a file
            file_path = os.path.join(DOCUMENTS_DIR, random_file)  # Create the full path
            print(f"📄 Selected file: {file_path}")

            # Process document: chunk it and store embeddings in Chroma vector store
            process_file(file_path)

            # Test Text-to-Speech with a sample text chunk
            sample_text = "పిల్లలూ, ఈరోజు మనం భూగోళం గురించి నేర్చుకుందాం!"  # Example in Telugu
            text_to_audio(sample_text, lang='te')  # Telugu TTS (change 'te' to 'hi' for Hindi, 'en' for English, etc.)
        else:
            print("❌ No valid documents found in the 'documents' directory.")
    except FileNotFoundError:
        print(f"❌ Directory '{DOCUMENTS_DIR}' not found.")
```
